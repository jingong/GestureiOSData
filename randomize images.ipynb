```python
from IPython.display import Image
import random
previous = ['bite.png', 'left_tap.png', 'right_tap.png', 'front_tap.png']
names = {
    'bite.png': "Bite with the back of your teeth",
    'left_tap.png': "Bite with your left side teeth",
    'right_tap.png': "Bite with your right side teeth",
    'front_tap.png': "Bite with the your front teeth"
}
```

```python
def randomize_images():
    image_names = ['bite.png', 'left_tap.png', 'right_tap.png', 'front_tap.png']
    global previous
    while image_names == previous:
        random.shuffle(image_names)
    previous = image_names    
    return image_names
```

```python
rand = randomize_images()
```

```python
print(rand)
```

```python
print(names[rand[0]])
```
